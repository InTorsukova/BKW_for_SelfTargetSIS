```python
import math

# Задаем начальные параметры
m=2*256
q=8380417
y1=523776 #γ_1
bb=375 #β
y2=y1/2 #γ_2
nu=7 #η
d=14
tau=round(bb/nu) #τ
be1=y1-bb
be2=y1+1+2^(d-1)*tau
if be1>be2:
    beta=be1
else:
    beta=be2
print('Входные параметры: m =',m, 'q =', q,'beta =', beta,'y2 =', y2,'tau =', tau, 'beta1 =', be1,'beta2 =', be2 )

# Считаем число шагов t
B=[]
B.append(beta)
print('B0 =', B[0])
Bt=beta
t_test = 1
while Bt<q: 
    Bt=Bt*2
    t_test+=1
Bt=Bt/2
print('Bt =', Bt)

t=log(Bt//B[0],2)+1
print('t = ',t, 't_test = ', t_test)

# Находим все B[i]
for i in range(1,t):
    B.append(2*B[i-1])
print('B =', B)

# Находим все m[i]
su=0
for a in range(t):
    su+=a/(log(q/B[0])-a*log(2))
mlist=[]
mlist.append(floor(m/t+log(2)*su))
print('m0 =',mlist[0])
mprov=m-mlist[0]
for i in range(1,t):
    mm=floor(mlist[0]*log(q/B[0])/log(q/B[i]))
    mprov-=mm
    mlist.append(mm)  
print('m =',mlist)

# Считаем сложность   
for i in range(1,t):
    print(i, mlist[i]*log(q/B[i]).n())
```
